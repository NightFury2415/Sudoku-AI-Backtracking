```python
sudoku_df = pd.DataFrame(pd.read_csv('../input/sudoku/sudoku.csv', nrows = 10**2))
```

```python
def shape(sudoku_df):
    for n in range(sudoku_df.shape[0]):
        sudoku_df.iloc[n,0] = np.reshape(list(sudoku_df.puzzle.values[n]),(9,9)).astype(int)
        sudoku_df.iloc[n,1] = np.reshape(list(sudoku_df.solution.values[n]),(9,9)).astype(int)
    return sudoku_df
```

```python
sudoku_df = shape(sudoku_df)
sudoku_df.iloc[0,0]
```

```python
grid = sudoku_df.iloc[0,0]
```

```python
def possible(row, column, number):
    global grid
    #Is the number already existing in the given row?
    for i in range(0,9):
        if grid[row][i] == number:
            return False
        
    #Is the number already existing in the given column?
    for i in range(0,9):
        if grid[i][column] == number:
            return False
    
    #Is the number already existing in the given square?
    x0 = (column // 3) * 3
    y0 = (row // 3) * 3
    for i in range(0,3):
        for j in range(0,3):
            if grid[y0+i][x0+j] == number:
                return False
    #If not then it is a possible solution to that space
    return True
```

```python
def solve():
    global grid
    for row in range(0,9):
        for column in range(0,9):
            if grid[row][column] == 0:
                for number in range(1,10):
                    #check is the number is a valid value
                    if possible(row, column, number):
                        #if the number is valid for the space/block put it in our space/block
                        grid[row][column] = number
                        #check for validity or other solutions
                        solve()
                        #if the number isn't valid then reset or empty the space/block
                        grid[row][column] = 0

                return
      
    print(np.matrix(grid))
    print('\nCheck for other possible solutions?')

solve()
```
